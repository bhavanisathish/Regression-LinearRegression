```python
import matplotlib.pyplot as plt
import numpy as np


class regression():
    def __init__(self):
        pass
    def predict(self,X):
        y = self.bias + self.theta *X
        
        return y
    def compute(self,X,theta,bias):
        y = bias + theta *X
        return y
    def computecost(self,X,Y,theta,bias):
        error=0.0
        m = len(X)
        for i in range(m):
            p = self.compute(X[i],theta,bias)
            error +=  (Y[i] - p)**2
        return error/m
    def gradient(self,X,Y,theta,bias,alpha):
        m=len(X)
        theta_derivative =0
        bias_derivative=0

        for i in range(m):
            p = self.compute(X[i],theta,bias)
            theta_derivative += - 2*X[i]*(Y[i] -p ) 
            bias_derivative += - 2*(Y[i] -p )

        theta -= theta_derivative * alpha /m
        bias -=bias_derivative * alpha/m

        return theta , bias
    
    def train(self,X,Y,theta,bias,alpha,iteration):
    
        cost_hist = []

        for i in range(iteration):

            theta,bias = self.gradient(X,Y,theta,bias,alpha)

            cost = self.computecost(X,Y,theta,bias)
            cost_hist.append(cost)
            

       # tup = self.r2_score(test_y, bias,theta,test_x)
        #print(tup)
        self.bias =bias
        self.theta =theta
        print('intercept',self.bias)
        print('coef',self.theta)      
        costs = np.squeeze(cost_hist)
        plt.plot(costs)
        plt.ylabel('cost')
        plt.xlabel('iterations (per hundreds)')
        plt.show()
        


        return theta,bias,cost_hist
    def r2_score(self,y_true,X):
        '''
            r2 = 1 - (rss/tss)
            rss = sum_{i=0}^{n} (y_i - y_hat)^2
            tss = sum_{i=0}^{n} (y_i - y_bar)^2
        '''
        y_pred = self.bias + self.theta *X
        
        y_average = np.average(y_true)

        residual_sum_of_squares = 0
        total_sum_of_squares = 0

        for i in range(len(y_true)):
            residual_sum_of_squares += (y_true[i] - y_pred[i])**2
            total_sum_of_squares += (y_true[i] - y_average)**2
            
        
        

        return 1 - (residual_sum_of_squares/total_sum_of_squares)
    
    
```
