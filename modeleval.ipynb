```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
ps = pd.read_csv('FuelConsumption.csv')
```

```python
data = np.random.randn(len(ps))<0.8
trainset = ps[data]
testset = ps[~data]
```

```python
train_x = trainset['ENGINESIZE'].values
train_y =trainset['CO2EMISSIONS'].values

test_x = testset['ENGINESIZE'].values
test_y =testset['CO2EMISSIONS'].values
```

```python

theta = 0
bias = 0
alpha = 0.01
iters = 1500
reg = regression()
reg.train(train_x,train_y,theta,bias,alpha,iters)
```

```python
r22t =reg.r2_score(train_y,train_x)
r2 = reg.r2_score(test_y,test_x)
predict = reg.predict(test_x)

plt.scatter(train_x,train_y)
plt.plot(test_x,predict,color='red')
plt.show()
       
print('Train accuracy',r22t*100,' %')
print('Test accuracy',r2*100,' %')
```
